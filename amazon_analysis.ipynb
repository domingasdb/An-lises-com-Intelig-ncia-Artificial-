```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('amazon.csv')
data
```

```python
data.info()
```

```python
data[data['rating']!='|']['rating'].astype(float).mean()
```

```python
data['rating'] = data['rating'].str.replace('|','4.09')
```

```python
data['rating'] = data['rating'].astype(float)
```

```python
data['discounted_price'] = data['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
data['actual_price'] = data['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
data['percent'] = (data['discounted_price']-data['actual_price'])/data['actual_price']
data
```

```python
from forex_python.converter import CurrencyRates

taxa_de_cambio = CurrencyRates().get_rate('INR', 'BRL')

data['real_converted_price'] = data['actual_price'] * taxa_de_cambio
data
```

```python
# Function to classify 'actual_price'
def classify_price(actual_price):
    if actual_price < 50:
        return 'less than 50'
    elif 50 <= actual_price <= 500:
        return 'between 50 and 500'
    else:
        return 'more than 500'

# Apply the classification function to create a new column
data['price_classification'] = data['actual_price'].apply(classify_price)
data['price_classification'].value_counts()
```

```python
classification_counts = data.groupby('price_classification').size().reset_index(name='count')
classification_counts
```

```python
def classify_rating(rating):
    if rating <= 2:
        return 'Terrible'
    elif 2 < rating <= 3:
        return 'Poor'
    elif 3 < rating <= 4:
        return 'Fair'
    elif 4 < rating <= 4.5:
        return 'Good'
    else:
        return 'Excellent'

data['rating_classification']=data['rating'].apply(classify_rating)
```

```python
classification_counts = data.groupby('rating_classification').size().reset_index(name='count')
classification_counts
```

```python
category = data['category'].str.split('|', expand=True)
category
```

```python
category.info()
```

```python
data['category0'] = category[0]
data['category1'] = category[1]
data['category3'] = category[2]
```

```python
pd.pivot_table(data[['category0','category1','product_id']], index=['category0','category1'], aggfunc='count')
```

```python
data_clean = data[['price_classification','actual_price', 'discounted_price', 'rating_classification','category0', 'category1','percent']]
data_clean['percent'] = data_clean['percent'].astype('str')
```

```python
data_clean.to_csv('data_clean.csv')
```
